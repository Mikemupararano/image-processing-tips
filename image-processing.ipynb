```python
#Drawing images
#Create a black image
image_shape=(600, 600, 3)
black_image=np.zeros(shape=image_shape, dtype=np.int16)
#Show the image
plt.imshow(black_image)
```

```python
#Create a blank image with a white background
image_shape=(600, 600, 3)
white_image=np.zeros(shape=image_shape, dtype=np.int16)
#Set every pixel of the image to 255
white_image.fill(255)
#Show the image
plt.imshow(white_image)
```

```python
#Draw a line on black image
line=cv2.line(black_image, (599,0), (0,599), (0,255,0), 4)
#Show image
plt.imshow(line)
```

```python
#Drawing a circle
#Create white image
img_shape=(600, 600, 3)
white_image=np.zeros(shape=image_shape, dtype=np.int16)
#Set every pixel of the image to 255
white_image.fill(255)
#Draw a red circle on white image
circle=cv2.circle(white_image, (300, 300), 100, (255, 0, 0 ), 6)
#Show image
plt.imshow(circle)
plt.show()
```

```python
#Drawing a rectangle on a black image
#Create a black image
image_shape=(600, 600, 3)
black_image=np.zeros(shape=image_shape, dtype=np.int16)
#Draw a green rectangle on a dark image
rectangle=cv2.rectangle(black_image, (200, 200), (400, 500), (0, 255, 0), 5)
#Show the image
plt.imshow(rectangle)
plt.show()
```

```python
#Fill the reangle with a specified color
#Create a black image
image_shape=(600, 600, 3)
black_image=np.zeros(shape=image_shape, dtype=np.int16)
black_image=cv2.rectangle(black_image, (200,200), (400, 500), (0, 255, 0), -1)
#Draw the image
plt.imshow(black_image)
plt.show()
```

```python
#Writing on images
#Create a black image
image_shape=(600, 600, 3)
black_image=np.zeros(shape=image_shape, dtype=np.int16)
#Write on black image
text=cv2.putText(black_image, 'Thanksgiving', (10, 500),
                cv2.FONT_HERSHEY_SIMPLEX, 3, (255, 0, 0), 2, cv2.LINE_AA)
#Display image
plt.imshow(text)
```

Flipping images


```python
#Read image
image=cv2.imread('messi.png')
```

```python
#Convert image color space BGR to RGB
rgb_image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
#Display image
plt.imshow(rgb_image)
plt.axis('off')  # Hide axes for better display
plt.show()
```

```python
#Flip image horizonatally using flip(rgb_image, 1) function. The argument 1 flips the image horizontally.
image_flip=cv2.flip(rgb_image, 1)
#Display image
plt.imshow(image_flip)
plt.axis('off')
plt.show()
```

```python
#Flip image vertically by passing the argument 0
image_flip=cv2.flip(rgb_image, 0)
#Display image
plt.imshow(image_flip)
plt.axis('off')
plt.show()
```

```python
#Flip image on both axis by passing argument -1 to the flip() function
image_flip=cv2.flip(image_flip, -1)
#Draw image
plt.imshow(image_flip)
#plt.axis('off')
plt.show()
```

```python
#Changing the brightness of images
#Read image
image=cv2.imread('nature.jpeg')
#Conevert BGR color space to RGB
rgb_image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#Display the image
plt.imshow(rgb_image)
plt.show()
```

```python
#Change brightness of image
#Set weightage for alpha  and beta
alpha_=1
beta_=50
#Add weight to the original image to change the brightness
image_change=cv2.addWeighted(rgb_image, alpha_, 
                             np.zeros(image.shape, image.dtype), 0, beta_)
#Display the image
plt.imshow(image_change)
plt.show()
```
